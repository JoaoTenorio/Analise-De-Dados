```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("disney_movies_total_gross.csv")
```

```python
df.head()
```

```python
df = df.rename(columns={'movie_title': 'Titulos Dos Filmes', 'release_date': 'Data de Lançamento', 'genre': 'Genero',
                        'mpaa_rating': 'Avaliação MPAA', 'total_gross': 'Total Bruto', 'inflation_adjusted_gross': 'Inflação Bruta Ajustada'})
```

```python
# converting to datetime
df['Data de Lançamento'] = pd.to_datetime(df['Data de Lançamento'])
```

```python
df.info()
```

```python
df['Ano'] = df['Data de Lançamento'].dt.date.astype(str).str.split('-', expand=True)[0]
```

```python
df.head()
```

```python
# top 5 movies with best total_gross
df.sort_values('Total Bruto',ascending=False).head(5)
```

```python
# top 5 movies with best inflation_adjusted_gross
df.sort_values('Inflação Bruta Ajustada',ascending=False).head(5)
```

```python
import seaborn as sns
%matplotlib inline
```

```python
plt.style.use('dark_background')
```

```python
sns.countplot(y='Genero',data=df)
```

Filmes de comedia sao maioria

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=90)
sns.histplot(df['Ano'],kde=True)
```

Esse grafico mostra que os filmes em sua maioria foram lançados em 1994

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=90)
sns.histplot(x='Ano',hue='Genero',data=df)
```

```python
sns.barplot(y='Genero',x='Total Bruto',data=df)
```

Filmes de aventura e açao sao os que mais tem rentabilidade

Com a inflaçao ajustada, filmes musicais tiveram maior valor bruto
